```python
for i in xrange(1, len(files)):
#for i in xrange(1, 2):

    past = pd.read_csv('Queries/gym_team/' + files[i - 1])
    current = pd.read_csv('Queries/gym_team/' + files[i])
    
    df = pd.concat([past, current], axis=1)
    df.columns = col
    
    for j in xrange(0, df.shape[0]):
        if df.Gym_A[j] == df.Gym_B[j]:
            if df.Team_A[j] != rotacao[df.Gym_A[j]][0]:
                rotacao[df.Gym_A[j]][1] = rotacao[df.Gym_A[j]][1] + 1
                rotacao[df.Gym_A[j]][0] = df.Team_A[j]
```

```python
resultado = pd.DataFrame(rotacao.items(), columns=['Gym', 'Rotacao'])
```

```python
resultado.Rotacao = resultado.Rotacao.apply(lambda x : x[1]) 
```
