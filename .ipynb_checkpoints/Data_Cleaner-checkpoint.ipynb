```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
form_raw = pd.read_excel('pesquisa.xlsx', sheetname='Sheet1')
```

```python
new_column = ['Date', 'Team', 'Level', 'Age', 'Hour_Week', 'Hour_Weekend', 'Pokestops_Num', 'Gyms_Num', 'Pokestop_Neighbor', 
              'Gym_Neighbor', 'Raid_Num', 'Search_Pokemon', 'Badge_Gold', 'Badge_Silver', 'Badge_Bronze',
              'Coins', 'Gen_1', 'Gen_2', 'Gen_3', 'Raid_Ex', 'Raid_Ex_Loc'] 
```

```python
form_raw.columns = new_column
```

Alterar form_raw.Raid_Ex para numérico: <br>
Não = 0 <br>
Sim = 0 <br>
Sim, mais de uma vez = valores da Raid_Ex_Loc

```python
form_raw['Raid_Ex_Loc'] = pd.read_csv('RaidLoc.csv')
```

```python
form_raw['Raid_Ex'] = form_raw.Raid_Ex_Loc.apply(lambda x: str(x).count(';'))
```

```python
aux = pd.read_csv('Poke_Gym_Loc.csv')
aux.Pokestop_Neighbor = aux.Pokestop_Neighbor.apply(lambda lido: lido.replace(",",";"))
aux.Pokestop_Neighbor = aux.Pokestop_Neighbor.apply(lambda lido: lido + ";")
aux.Gym_Neighbor = aux.Gym_Neighbor.apply(lambda lido: lido.replace(",",";"))
aux.Gym_Neighbor = aux.Gym_Neighbor.apply(lambda lido: lido + ";")

form_raw.Pokestop_Neighbor = aux.Pokestop_Neighbor
form_raw.Gym_Neighbor = aux.Gym_Neighbor
```

```python
form_raw.Raid_Ex_Loc = form_raw.Raid_Ex_Loc.fillna("-")
```

```python
form_raw.Gen_1 = form_raw.Gen_1.fillna('Não')
form_raw.Gen_2 = form_raw.Gen_2.fillna('Não')
form_raw.Gen_3 = form_raw.Gen_3.fillna('Não')
```
