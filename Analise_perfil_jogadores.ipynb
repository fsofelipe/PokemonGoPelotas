```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
plt.rcParams['figure.figsize'] = (15,7)
```

```python
def autolabel(rects, ax):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()
        label_position = height + (y_height * 0.01)

        ax.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height),
                ha='center', va='bottom')
```

### Mystic

```python
mystic = form[form.Team == 'Mystic'].copy()
```

```python
mystic.Level.describe()
```

```python
mystic.nlargest(10, 'Level')
```

```python
ax = sns.kdeplot(mystic.Level, shade=True,bw=0.8)
```

```python
mystic.Age.describe()
```

```python
mystic.Hour_Week.describe()
```

```python
mystic.Hour_Weekend.describe()
```

```python

#ax1 = plt.subplot(131)

fig, ax = plt.subplots()

index = np.arange(3)
bar_width = 0.35
opacity = 0.8

s = mystic[mystic.Gen_1 == 'Sim'].Team.count()
n = mystic[mystic.Gen_1 == 'Não'].Team.count()
s1 = (s*100/(s+n))
n1 = (n*100/(s+n))

s = mystic[mystic.Gen_2 == 'Sim'].Team.count()
n = mystic[mystic.Gen_2 == 'Não'].Team.count()
s2 = (s*100/(s+n))
n2 = (n*100/(s+n))

s = mystic[mystic.Gen_3 == 'Sim'].Team.count()
n = mystic[mystic.Gen_3 == 'Não'].Team.count()

s3 = (s*100/(s+n))
n3 = (n*100/(s+n))

yes_bar = [s1, s2, s3]
no_bar = [n1, n2, n3]

rects1 = plt.bar(index, yes_bar, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Sim')

rects2 = plt.bar(index+bar_width+ 0.005, no_bar, bar_width,
                 alpha=opacity,
                 color='g',
                 label=u'Não')

autolabel(rects1, ax)
autolabel(rects2, ax)

plt.xticks(index + bar_width, (u'Geração 1', u'Geração 2', u'Geração 3'))
plt.xlabel(u'Gerações')
plt.ylabel('Porcentagem')
plt.title('Pokedex Completa')

plt.legend()
```
